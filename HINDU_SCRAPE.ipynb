```
import requests as r
import re
from bs4 import BeautifulSoup as soup
import pandas as pd
LINK=['https://www.thehindu.com/news/top-news-of-may-1-2020-lockdown-extended-by-two-more-weeks-shramik-special-trains-to-transport-stranded-migrant-workers-and-others-and-more/article31483375.ece']
regex = re.compile('.*content-body-.*')
CC=[regex]
n=len(LINK)
fdata=[]
for i in range(0,n):
  data=r.get(LINK[i])
  page_data=soup(data.text,'html.parser')
  State=page_data.find('a',{'class':'section-name'}).text
  Header=page_data.find('h1',{'class':'title'}).text
  dp=page_data.find('div',{'class':'ut-container'})
  pad=dp.find_all('span',{'class':'blue-color ksl-time-stamp'})
  if(len(pad)==2): 
    Place=pad[0].text
    Place = Place.replace(',', '')
    Date=pad[1].text
  elif(len(pad)==1):
    Place='India'
    Date=pad[0].text
  Content=page_data.find('div',{'id':CC[i]}).text
  column=[Date,State,Place,Header,Content]
  news=[]
  for line in column:
    news.append(line.strip().split('\n'))
    fdata.append(news)
df=pd.DataFrame(fdata)
for each_col in df.columns:
 df[each_col] =  df[each_col].apply(lambda x: x[0])
print(fdata)
```
